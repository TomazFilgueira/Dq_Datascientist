# Introduction

Deep learning algorithms use an artificial neural network, a computing system which learns high-level features from data by increasing the depth (i.e., number of layers) in the network. The inspiration for neural networks comes partially from biological neural networks. The cells in most brains (including ours) connect and work together. We call each of these cells in a neural network a neuron.

#Shallow Neural Networks

A neural network is comprised of the following components:

1. **Input Layer:** This is where the training observations are fed through the independent variables.

1. **Hidden Layers:** These are the intermediate layers between the input and output layers. This is where the neural network learns about the relationships and interactions of the variables fed in the input layer.

1. **Output Layer:** This is the layer where the final output is extracted as a result of all the processing which takes place within the hidden layers.

1. **Node:** A node, also called a neuron, in a neural network is a computational unit that takes in one or more input values and produces an output value.

A shallow neural network** is a neural network with a small number of layers--often one or two layers.** Shallow neural networks are typically used for simple tasks, such as regression or classification.

#Deep Neural Networks

In contrast to the shallow neural network, a deep (dense) neural network is composed of multiple hidden layers, each of which contains a set of neurons that learn to extract certain features from the data. The output layer produces the final results of the network.

The extra hidden layers in a deep neural network allow it to learn more complex patterns than a shallow neural network. As a result, deep neural networks are more accurate than shallow neural networks, but they are also more computationally expensive to train. For this reason, shallow neural networks are sometimes used when computational resources are limited. **For complex real time real world applications-**-such as multi-variate time series forecasting, natural language processing, real time forecasting, or predictive lead times--deep neural networks are preferable.



#Information flow

If you look at the image above, you'll see that the neural network flows in the forward direction.

 This process of passing input data through the layer(s) of a neural network to generate an output is called forward propagation.
 1. The input data is first fed into the input layer;

 1.which passes it to the hidden layer(s).

 1.The hidden layer(s) process the data and pass it to the output layer;

 1. Output layer produces the final output.

The calculations at the nodes are performed using the **multiply then sum** process. If you're familiar with vector algebra or linear algebra, that operation is commonly called a **dot product**. Let's understand this method of forward propagation by looking at the example below.

#Forward Propagation

For simplicity, we'll predict `Salary`, measured in thousands of USD, based on two variables: `age` and work `experience`, both measured in years. The diagram below shows the details for a professional who is 24 years old with 5 years work experience. The forward propagation algorithm will process this information through the network to make a prediction in the output layer.

In a neural network, lines connect the inputs to the hidden layer. The weight associated with each line indicates how strongly that input affects the hidden node where that line ends. The first step is to calculate the values in the nodes, n1 and n2, using the input values and the associated weights.

image.png

The calculation for node n1 is: $\displaystyle (24 \times 3) + (5 \times -4) = 52$.

Similarly, the calculation for node n2 is: $\displaystyle (24 \times 1) + (5 \times 2) = 34$.

image.png

Finally, this process is repeated for the intermediate hidden layers. Since we have only one hidden layer in this example, the next step will be to repeat this process for the final layer, which is the output layer. This calculation then becomes: $\displaystyle (52 \times 2) + (34 \times -1) = 70$. So, we predict a salary of $70,000 for this professional.

```python
input_layer = np.array([34, 7])
weights_topnode = np.array([2, -5])
weights_bottomnode = np.array([1, -2])
output_weights = np.array([3, -1])

top_node_value = np.sum(input_layer*weights_topnode)


bottom_node_value = np.sum(input_layer*weights_bottomnode)



hidden_layer_values = np.array([top_node_value, bottom_node_value])

prediction = np.sum(hidden_layer_values*output_weights)
print(top_node_value)
print(bottom_node_value)
print(prediction)
```

#Forward Propagation for Deeper Neural Network

The difference between deep and shallow neural networks is the use of many successive hidden layers. We forward propagate through these layers the same way we did for single hidden-layer models. The calculation method is similar, but it captures more context about those patterns present within the data, which makes prediction results much better!

```python
input_layer = np.array([34, 7])
weights_node_n1 = np.array([2, -5])
weights_node_n2 = np.array([1, -2])
weights_node_n3 = np.array([3, -2])
weights_node_n4 = np.array([1, -1])
output_weights = np.array([2, -6])

n1 = np.sum(input_layer*weights_node_n1)
n2 = np.sum(input_layer*weights_node_n2)

layer_1_output = np.array([n1,n2])
print(layer_1_output)

n3 = np.sum(layer_1_output*weights_node_n3)
n4 = np.sum(layer_1_output*weights_node_n4)
layer_2_output = np.array([n3,n4])
print(layer_2_output)

prediction = np.sum(layer_2_output*output_weights)
print(prediction)
```

#Activation Functions

So far, we've relied on the **multiply then sum** process to find the node values of the hidden layer. For neural networks to achieve their maximum predictive power through deep learning, we also use something called an **activation function** in the hidden layers to determine the final node values.

##Sigmoid Activation Function

The sigmoid function looks like an S-shaped curve:

image.png


You'll notice that the sigmoid function has horizontal asymptotes at 0 and 1, which means any input value will always output a value between 0 and 1. The sigmoid activation is an ideal activation function for a **binary classification problem** where the output is interpreted as a binomial probability distribution

##Softmax Activation Function

Softmax activation function is a type of sigmoid function that is commonly used in neural network architectures such as multi-layer perceptrons and convolutional neural networks. The possible output values of the Softmax activation function are **probabilities that range from 0 to 1.** The sum of all the probabilities should equal 1, as each value represents the probability of a certain class label being the correct one. For example, if an input is classified into 3 classes, then the softmax output will be three values between 0 and 1 where they add up to 1. While sigmoid is used for binary classification problems, the softmax activation function is better for multiclass classification tasks.

##ReLu Activation Function

ReLU is the most popular and widely-used activation function in the industry, as it is far more computationally efficient compared to the sigmoid and tanh functions. ReLU stands for **rectified linear unit** and is defined as follows:

$\displaystyle\text{ReLU}(x) =$

The `max(0,x) `function call returns the maximum value** between 0 and x**, which means that:

* when x is less than or equal to 0, the **value 0** is returned, and
* when x is greater than 0, the **value x** is returned.

##Tanh Activation Function

The tanh function is also known as the hyperbolic tangent function. It's similar to the sigmoid function, with an S-shaped curve. The tanh function takes any real value and outputs a value between -1 and +1. The larger your input value, the closer the output will get to 1; whereas the smaller the input value, the closer the output will be to -1. #Implementing ReLu Function

if we want to add non-linearity into the modeling process--that's where activation functions come into play! Since ReLU is the most popular activation function used in the industry, we'll learn how to implement ReLU activation in Python.

image.png

```python
input_layer = np.array([34, 7])
weights_node_n1 = np.array([2, -5])
weights_node_n2 = np.array([1, -5])
weights_node_n3 = np.array([3, -2])
weights_node_n4 = np.array([1, -1])
output_weights = np.array([2, -4])

def relu(input_value):
    output_value = max(0, input_value)
    return(output_value)

n1 = sum(input_layer*weights_node_n1)
n1_output = relu(n1)
print(n1_output)

n2 = sum(input_layer*weights_node_n2)
n2_output = relu(n2)
print(n2_output)

layer_1_output = np.array([n1_output,n2_output])
print(layer_1_output)

n3 = sum(layer_1_output*weights_node_n3)
n3_output = relu(n3)
print(n3_output)

n4 = sum(layer_1_output*weights_node_n4)
n4_output = relu(n4)
print(n4_output)

layer_2_output = np.array([n3_output, n4_output])
print(layer_2_output)

prediction= sum(layer_2_output*output_weights)
print(prediction)
```
