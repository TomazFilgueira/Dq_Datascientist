#The Tensor

**A tensor** is an array of data that can be processed by TensorFlow. A tensor can be represented as a matrix or a vector. In simple terms, we can think of it as a collection of numbers arranged into a particular shape. Mathematically, a tensor can be used to create n-dimensional datasets.

A zero-dimensional tensor is a **scalar**, which contains a single value and has no axes. A one-dimensional tensor is a **vector** which contains a list of values and has one axis. A two-dimensional tensor is a **matrix** that contains values stored across two axes.

We can create a basic tensor object consisting only of ones with the **tf.ones()** function, like this:

    t0 = tf.ones((3,))
    print(t0)
    Output
    tf.Tensor([1. 1. 1.], shape=(3,), dtype=float32)

The shape argument in the above output shows that t0 is a one-dimensional tensor where the value 1 is repeated three times. We can change the value of shape to create two-dimensional tensors.

The command below creates a two-dimensional tensor with **one row and three columns**.

    t1 = tf.ones((1,3))
    print(t1)
    Output
    tf.Tensor([[1. 1. 1.]], shape=(1, 3), dtype=float32)

If we want to create a tensor that contains different values, we can use the **tf.range()** function, like this:

    t2 = tf.range(start=1, limit=15, delta=2)
    print(t2)
    Output
    tf.Tensor([ 1  3  5  7  9 11 13], shape=(7,), dtype=int32)

```python
import tensorflow as tf
#tensor of ones - shape(1,7)
x1 = tf.ones((1,7))

#tensor of zeros - shape(1,3)
x2 = tf.zeros((1,3))

#tensor of odd numbers from 21 to 30
x3 = tf.range(start=21,limit=30,delta=2)

print(x1)
print(x2)
print(x3)
```

#Define Tensor as Constants

A constant tensor does not change, nor can it be trained. It can, however, have any dimension.

We can create a basic constant tensor object with the `tf.constant()` function. As an example, the code below creates a one-dimensional constant tensor from a python list.

    c1 = tf.constant([1, 2, 3, 4, 5, 6])
    print(c1)
    Output
    tf.Tensor([1 2 3 4 5 6], shape=(6,), dtype=int32)

The output above corresponds with the following parameters that we can pass to the `tf.constant()` function:

* `value`: a list of constant values for the dtype output.
* `shape`: defines the dimensions of the resulting tensor.
* `dtype`: defines the data type of the resulting tensor.

```python
import tensorflow as tf


c2 = tf.constant([100, 200, 300, 400, 500, 600], shape=[2, 3])
print(c2)

#Create a two by three constant tensor consisting only of the value 2
c3 = tf.constant(2, shape=[2, 3])
print(c3)


#Create a constant tensor that stores a unit matrix of order 3 as values.
c4 = tf.constant([1, 0, 0, 0, 1, 0, 0, 0, 1], shape=[3, 3])
print(c4)

#change the type to float32
c5 = tf.constant([100, 200, 300, 400, 500, 600], shape=[2, 3], dtype=tf.float32)
print(c5)
```

#Define Tensor as Variables

Unlike a constant tensor, a variable tensor's value can change during computation. Variables are the recommended way to represent shared and persistent states in order to store and manipulate data in TensorFlow. However, data type and shape are fixed.

We can create a basic variable object with the `tf.Variable` class. As an example, the code below creates a one-dimensional variable:

    v1 = tf.Variable([1, 2, 3, 4, 5, 6])
    print(v1)
    Output
    tf.Variable 'Variable:0' shape=(6,) dtype=int32, numpy=array([1, 2, 3, 4, 5, 6], dtype=int32)

The biggest difference between constants and variables is that a tf.Variable object represents a tensor whose value can be changed. This can be done using the `assign()` method, with the following syntax:
    
    variablename.assign(value=new_values)

```python
v2 = tf.Variable([True, True, False, False])
print(v2)

v3 = tf.Variable([[20,25],[22,16]])
print(v3)

v3.assign(value=[[15,20],[12,16]])
print(v3)
```

#Convert Tensors to Numpy Array and Vice-Versa

Tensors are similar to NumPy arrays, with the key difference between them being that tensors have accelerator support like GPUs (graphics processing units) and TPUs (tensor processing units). These processors provide much better performance when it comes to the thousands (or millions) of calculations involved in deep learning models, compared to a CPU (central processing unit).

It's easy to convert tensors to NumPy arrays using the numpy() method, like this:

    array_name = tensor_name.numpy() or
    array_name = np.array(tensor_name)

The above commands show the ease with which tensor objects can be converted into NumPy arrays. At the same time, it's easy to convert NumPy arrays to tensor objects. This can be done with the `tf.convert_to_tensor()` function


```python

#Define a variable tensor n1
n1 = tf.Variable([[10,15],[20,25]])
print(type(n1))

#Convert n1 into a numpy array with the tensor_name.numpy()
array1 = n1.numpy()
print(type(array1))

#Convert n1 into a NumPy array with the numpy.array()
array2 = np.array(n1)
print(type(array2))

#Convert the NumPy object array2 back to a tensor object
nt1 = tf.convert_to_tensor(array2)
print(type(nt1))
```

#Basic Mathematical Operations on Tensors

TensorFlow provides several functions to perform mathematical operations on tensors. Each operation is performed with a tf function, and each operation returns a tensor.

The following list contains the basic mathematical operations that are most often performed:

* `tf.add(x, y): `performs addition on two tensors x and y.

* `tf.subtract(x, y):` performs subtraction on two tensors x and y.

* `tf.multiply(x, y):` computes the product of two tensors x and y.

* `tf.divide(x, y):` divides the elements of two tensors x and y.

```python
t1 = tf.constant([6, 4, 10])
t2 = tf.constant([3, 2, 4])

t3 = tf.Variable([[20, 25], [22, 16]])
t4 = tf.Variable([[10, 5], [11, 8]])

op1 = tf.add(t1,t2)
print(op1)

op2 = tf.subtract(t3,t4)
print(op2)

op3 = tf.multiply(t1,t2)
print(op3)

op4 = tf.divide(t3,t4)
```

Good sum:

    t3 = tf.constant([0, 5, 11])

    ex2 = tf.add(t3, t4)

    tf.Tensor([2 7 13], shape=(3,), dtype=int32)

##Broadcasting

In its simplest terms, during an element-wise operation, broadcasting will attempt to stretch the smaller tensor to have the same shape as the larger one.

TensorFlow determines if two tensors are broadcast-compatible by aligning the last value of their shapes and checking that each pair of aligned dimensions meet either of the following conditions:

* The aligned dimensions are equal.

* One of the aligned dimensions is equal to 1.

```python
t5 = tf.constant([[2, 0], [3, 0], [1, 5], [0, 4]])
t6 = tf.constant([[3, 5], [8, 13]])
t7 = tf.constant([[1], [1], [4], [2]])

print(t5.shape)
print(t6.shape)
print(t7.shape)

#t5 can sum t7


op5 = tf.add(t5,t7)
```

#Advanced Mathematical Operations on Tensors

 The tf.math module in TensorFlow provides many functions for performing such calculations.

Some of the important mathematical operations often required in machine learning workflows, and their associated tf.math functions, are listed below:

* `Logarithmic Transformation:` when the data contains outliers or exhibits skewness, logarithmic transformation is often required. The `tf.math.log(x, name=None)` function performs this operation on the tensor x.

* `Missing Value`: identification of missing values is one of the first data exploration tasks in analytics. In TensorFlow, this can be done using the `tf.math.is_nan(x, name=None)` function.

* `Absolute Function`: when we're dealing with metrics such as mean absolute deviation or mean absolute percent error, we have to calculate the absolute value. This can be done with the `tf.math.abs(x, name=None)` function, where x represents a tensor. An example of this is shown below with the code and its output.

    x = tf.constant([-2.25, -48.23, 6.25])
    print(tf.math.abs(x))
    output:
    tf.Tensor([ 2.25 48.23  6.25], shape=(3,), dtype=float32)

 [Here](https://www.tensorflow.org/api_docs/python/tf/math#functions_2) are some popular mathematical operations you may need to use in the exercises below.



```python
t1 = tf.constant([6, 4, 10])
t2 = tf.constant([3, 2, 4])

t3 = tf.Variable([[2.4, 2.5], [2.2, 1.6]])
t4 = tf.Variable([[9, 5], [16, 8]])

t_log = tf.math.log(tf.cast(t1,tf.float32),name=None)

print(t_log)

t_cumsum = tf.cumsum(t2,name=None)
print(t_cumsum)

t_tanh = tf.tanh(t3,name=None)
print(t_tanh)


t_sqrt = tf.math.sqrt(tf.cast(t4,tf.float32),name=None)
print(t_sqrt)
```
