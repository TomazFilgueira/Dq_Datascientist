# 1) Finding Unique Values with SELECT DISTINCT

In this lesson we'll focus on filtering data that is **text-based**. Filtering on text values is important because many databases store a large amount of textual data, such as customer names, addresses, and product information. Text-based filtering allows us to find records that contain a certain word or phrase or match a specific pattern, even if we don't know the exact value to search for.

To get started, let's look at a useful way to find a list of all *distinct* values in a field.

This can be done in SQL with SELECT DISTINCT. For example, we can find all the unique values of category in the orders table:

```python
%%sql
SELECT DISTINCT category
  FROM orders;
```

You may have noticed that `subcategory` relates to `category`. To confirm, we could include both fields in our `SELECT DISTINCT` clause to find all the unique combinations of these categories. Notice that `DISTINCT` has to come before any column names and then applies to the whole query.

```python
%%sql
SELECT DISTINCT category, subcategory
  FROM orders;
```

# 2) Filtering for Categories with IN

In an earlier lesson we used the `IN` operator to find a group of non-consecutive numeric values in a field.

This operator can similarly be used with **text**, or **string**, data to pull out a list of categories in a field. 

For example, let's say we want to conduct an analysis of just the '`West`' and '`East`' regions of the orders table:

```python
%%sql
SELECT order_id, product_id, region
  FROM orders
 WHERE region IN ('West', 'East')
 Limit 5;
```

# 3) Flipping the Script with NOT

Keyword `NOT` can be used in other cases, too. Rather than keeping conditions that do meet some criteria, we keep those that do not.

For example, if we wanted to keep all records except those from the West region, we could write this query:

```python
%%sql
SELECT order_id, product_id, sales
  FROM orders
 WHERE NOT region = 'West'
 LiMIT 5;
```

The query above can also be written with the mathematical operator `<>`:

```python
%%sql
SELECT order_id, product_id, sales
  FROM orders
 WHERE region <> 'West'
 LiMIT 5;
```

Using `NOT` is more efficient when we're excluding a large number of values. This is because the database can use an index to quickly identify the records to exclude. In contrast, using the inequality operator requires scanning the entire table and comparing each row to the specified value.

While database efficiency is important, the true power of `NOT` occurs when we combine it with `IN`, because it allows us to exclude records that match any of the specified values in a list.

```python
%%sql
SELECT order_id, product_id, sales
  FROM orders
 WHERE region NOT IN ('West', 'South')
 Limit 5;
```

`NOT` can also be combined with other conditions using the AND and OR operators. This allows us to create more complex filtering conditions.

For example, let's say we want to keep records with sales greater than 500 and exclude those from the '`West`' and '`South`' regions:

```python
%%sql
SELECT order_id, product_id, sales
  FROM orders
 WHERE sales > 500 AND region NOT IN ('West', 'South')
 Limit 5;
```

# 4) Searching for Text Patterns with LIKE (Part 1)

SQL includes a number of so-called wildcard operators which are special characters that can be substituted for other characters.

For example, the `*` in `SELECT *` is a wildcard for "all."

The `LIKE` operator in conjunction with `WHERE` can be used to search for a specified pattern in a text column. In particular, we'll often see the `%` character, which matches any number of characters (including zero).

For example, Superstore carries a number of products from Xerox. Let's say we want to pull every record from the `product_name` column that starts with **Xerox**. We can do so by adding a `%` after Xerox to indicate SQL should return any record matching this pattern:

```python
%%sql
SELECT DISTINCT product_name
  FROM orders
 WHERE product_name LIKE 'Xerox %'
 LiMIT 5;
```

The % doesn't have to be used just once or at the end of words, though. For example, the query below shows matches for customer names that have the text "allen" anywhere.

Notice that Alejandro Ballentine was also included in the output because their last name contains the phrase "allen." This helps demonstrate that LIKE is not case sensitive.


```python
%%sql
SELECT DISTINCT customer_name
  FROM orders
 WHERE customer_name LIKE '%Allen%'
 LiMIT 5;
```

# 5) Searching for Text Patterns with LIKE (Part 2)

Another useful wildcard to use with LIKE is `_`. The `_` wildcard is a pattern matching character that matches any single character in a string. When used in conjunction with the LIKE keyword, it allows us to search for records that match a certain pattern, even if we don't know the exact value to search for.

For example, suppose we want to find all states in the orders table with names that have exactly **four characters**. We can use the `_` wildcard four times with the LIKE keyword to find all records that have a name with four characters. Here's an example query:

```python
%%sql
SELECT DISTINCT state
  FROM orders
 WHERE state LIKE '____';
```

The `_` wildcard can also be used in combination with `%`. For example, to find all customers whose first name starts with "M" and is exactly five characters long, we can write the following query.

Notice that "M" is followed by 4 _ wildcards to create a five-letter first name, and the `%` wild card happens after a space to account for last names of any length.

```python
%%sql
SELECT DISTINCT customer_name
  FROM orders
 WHERE customer_name LIKE 'M____ %';
```

# 6) Parentheses for Multiple Criteria

Earlier in the course we learned about selecting multiple criteria with `AND`, `OR`, `IN` and `BETWEEN`. These logical operators can get quite complex. Grouping conditions with parentheses can establish an order of operations inside our `WHERE` clause. Parentheses can also help add clarity to our query to improve readability.

For example, we've been asked to pull records of high-value transactions where sales are at least **$100**, or where units sold are *both* below five and profit is greater than 50.

In this case, we're using `AND` with `OR` statements together and want to explicitly define which go together:

```python
%%sql
SELECT order_id, product_id, sales, quantity, profit
  FROM orders
 WHERE sales > 100 OR
       (quantity < 5 AND profit > 50)
       
limit 5;
```
